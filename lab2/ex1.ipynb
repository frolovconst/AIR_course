```python
def sumDictValues(dctnr):
    N = 0
    for val in dctnr.values():
        N = + N + val
    return N
```

```python
def calc_probs(dctnr):
    prob_dctnr = dict()
    tot_size = sumDictValues(dctnr)
    for key in dctnr.keys():
        prob_dctnr[key] = dctnr[key]/tot_size
    return prob_dctnr
```

```python
prob_corpus = calc_probs(vocab)
```

```python
N_topics = len(dataset)
```

```python
def H_of_dict(dctnr):
    H = 0
    prob_dict = calc_probs(dctnr)
    for key in prob_dict.keys():
        H = H - prob_dict[key]*np.log2(prob_dict[key])
    return H/N_topics
```

```python
def H_of_dataset(dctnr):
    H = 0
    for val in dctnr.values():
        H = H + H_of_dict(val)
    return H
```
